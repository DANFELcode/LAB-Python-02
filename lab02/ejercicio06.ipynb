## Ejercicio 6 - Minimax

```python
import math

class NIM:
    def __init__(self, piedras_iniciales=10):
        self.piedras_iniciales = piedras_iniciales
        
    def es_terminal(self, piedras):
        """Verifica si el juego ha terminado"""
        return piedras <= 0
    
    def utilidad(self, piedras, turno_max):
        """
        Calcula la utilidad del estado terminal
        Si es terminal y pierde MAX -> -1
        Si es terminal y pierde MIN -> 1
        """
        if piedras <= 0:
            # Si es turno de MAX y no puede mover, perdió MAX
            return -1 if turno_max else 1
        return 0
    
    def movimientos_posibles(self, piedras):
        """Genera los movimientos posibles (1, 2 o 3 piedras)"""
        return [i for i in range(1, 4) if i <= piedras]
    
    def aplicar_movimiento(self, piedras, movimiento):
        """Aplica un movimiento y devuelve el nuevo estado"""
        return piedras - movimiento
    
    def minimax(self, piedras, es_turno_max):
        """
        Algoritmo Minimax
        Retorna: (mejor_valor, mejor_movimiento)
        """
        # Caso base: estado terminal
        if self.es_terminal(piedras):
            return self.utilidad(piedras, es_turno_max), None
        
        if es_turno_max:
            mejor_valor = -math.inf
            mejor_mov = None
            
            for mov in self.movimientos_posibles(piedras):
                nuevo_estado = self.aplicar_movimiento(piedras, mov)
                valor, _ = self.minimax(nuevo_estado, False)
                
                if valor > mejor_valor:
                    mejor_valor = valor
                    mejor_mov = mov
                    
            return mejor_valor, mejor_mov
        else:
            mejor_valor = math.inf
            mejor_mov = None
            
            for mov in self.movimientos_posibles(piedras):
                nuevo_estado = self.aplicar_movimiento(piedras, mov)
                valor, _ = self.minimax(nuevo_estado, True)
                
                if valor < mejor_valor:
                    mejor_valor = valor
                    mejor_mov = mov
                    
            return mejor_valor, mejor_mov
    
    def jugar(self):
        """Simula el juego completo"""
        piedras = self.piedras_iniciales
        turno_humano = True  # MIN = Humano, empieza primero
        
        print(f"\n=== JUEGO NIM ===")
        print(f"Piedras iniciales: {piedras}")
        print("Reglas: Puedes quitar 1, 2 o 3 piedras por turno")
        print("Pierde quien no pueda mover (sin piedras)")
        print("MAX = Computadora, MIN = Humano")
        print("Comienza el Humano (MIN)\n")
        
        while not self.es_terminal(piedras):
            print(f"\nPiedras restantes: {piedras}")
            
            if turno_humano:
                # Turno del humano (MIN)
                print("\n--- Tu turno (MIN) ---")
                
                while True:
                    try:
                        movimiento = int(input(f"¿Cuántas piedras quitas (1-{min(3, piedras)}): "))
                        if movimiento in self.movimientos_posibles(piedras):
                            break
                        else:
                            print(f"Movimiento inválido. Debe ser 1, 2 o 3 y no superar {piedras}")
                    except ValueError:
                        print("Por favor ingresa un número válido")
                
                piedras = self.aplicar_movimiento(piedras, movimiento)
                print(f"Quitaste {movimiento} piedras. Quedan {piedras}")
                
            else:
                # Turno de la computadora (MAX)
                print("\n--- Turno de la computadora (MAX) ---")
                valor, movimiento = self.minimax(piedras, True)
                
                if movimiento is None:
                    print("La computadora no puede mover")
                    break
                    
                piedras = self.aplicar_movimiento(piedras, movimiento)
                print(f"La computadora quitó {movimiento} piedras. Quedan {piedras}")
            
            # Cambiar turno
            turno_humano = not turno_humano
        
        # Determinar ganador
        print(f"\n=== FIN DEL JUEGO ===")
        print(f"Piedras restantes: {piedras}")
        
        if piedras <= 0:
            if turno_humano:
                # Si terminó en turno del humano, perdió el humano
                print("¡No puedes mover! ¡Has perdido!")
                print("Ganó la COMPUTADORA (MAX)")
            else:
                print("¡La computadora no puede mover! ¡Has ganado!")
                print("Ganó el HUMANO (MIN)")


# Ejecutar juego
if __name__ == "__main__":
    juego = NIM(piedras_iniciales=10)
    juego.jugar()
```
