## Realizar los siguientes ejercicios usando filter.

### a. Dada una lista de palabras, filtrar solo aquellas con mas de 4 letras.

```python
palabras = ["peru", "python", "semestre", "ok"]
print(list(filter(lambda p: p if len(p) > 4 else None, palabras)))
```

### b. Dada una lista filtrar los elementos que son nulos (None).

```python
elementos = [1, None, "hola", None, 45, None, "mundo", 0]
filtradas = list(filter(None, elementos))
print(filtradas) 
```

### c. Dada una lista de palabras, filtrar aquellas que empiezan pro una vocal.

```python
palabras = ["hola", "ana", "elefante", "casa", "isla", "oso", "arbol", "perro"]
vocales = "aeiouAEIOU"
filtradas = list(filter(lambda p: p and p[0] in vocales, palabras))
print(filtradas)
```

### d. Dada una lista de palabras, filtrar aquellas que son palindromos.

```python
palabras = ["radar", "hola", "reconocer", "python", "oso", "mundo", "anilina"]
filtradas = list(filter(lambda p: p == p[::-1], palabras))
print(filtradas)
```

### e. Dada una lista de numeros, filtrar los que terminan en 5.

```python
numeros = [15, 23, 35, 42, 55, 68, 75, 80, 95, 100]
filtradas = list(filter(lambda n: n % 10 == 5, numeros))
print(filtradas)
```

### f. Dada una lista de diccionarios, filtrar los productos con precio mayor a 100.

```python
productos = [
    {"nombre": "Teclado", "precio": 80},
    {"nombre": "Mouse", "precio": 40},
    {"nombre": "Monitor", "precio": 300},
    {"nombre": "Webcam", "precio": 150}
]

filtrados = list(filter(lambda p: p["precio"] > 100, productos))
print(filtrados)
```
