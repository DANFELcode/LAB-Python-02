## Ejercicio 4 - Sorted

### a. Ordenas una lista de palabras alfabeticamente.

```python
palabras = ["zebra", "manzana", "perro", "casa", "elefante"]
ordenadas = sorted(palabras)
print(ordenadas)  # ['casa', 'elefante', 'manzana', 'perro', 'zebra']
```

### b. Ordenar tuplas por el primer elemento

```python
tuplas = [(3, "tres"), (1, "uno"), (4, "cuatro"), (2, "dos")]
ordenadas = sorted(tuplas)
print(ordenadas) 
```

### c. Dada una lista de diccionarios, ordenar por precio (mayor a menor), y luego por nombre


```python
productos = [
    {"nombre": "Camisa", "precio": 30},
    {"nombre": "Pantalón", "precio": 55},
    {"nombre": "Medias", "precio": 10},
    {"nombre": "Chaqueta", "precio": 80},
    {"nombre": "Sombrero", "precio": 30}  # Mismo precio que Camisa
]

# Ordenar por precio (mayor a menor) y luego por nombre alfabéticamente
ordenados = sorted(productos, key=lambda x: (-x["precio"], x["nombre"]))
print(ordenados)
```

### d. Dada una lista de palabras, ordenarlas según su longitud.

```python
palabras = ["python", "java", "c", "javascript", "go", "rust"]
# De menor a mayor longitud
ordenadas = sorted(palabras, key=len)
print(ordenadas)  # ['c', 'go', 'java', 'rust', 'python', 'javascript']

# De mayor a menor longitud
ordenadas_desc = sorted(palabras, key=len, reverse=True)
print(ordenadas_desc)  # ['javascript', 'python', 'java', 'rust', 'go', 'c']
```

### e. Dada una lista de numeros, ordenarlos de tal forma que aparezcan primero los pares y luego los impares.

```python
numeros = [5, 2, 8, 3, 1, 6, 9, 4, 7]


ordenados = sorted(numeros, key=lambda x: (x % 2 != 0, x))
print(ordenados)

def clave_orden(x):
    return (x % 2 == 1, x)  # True para impares, False para pares

ordenados = sorted(numeros, key=clave_orden)
print(ordenados)  # [2, 4, 6, 8, 1, 3, 5, 7, 9]
```

### f. Convertir una palabra a un diccinario que cuente la aparicion de cada letra, ordenar dicho diccionario según la frecuencia de aparicion.




```python
from collections import Counter

palabra = "programacion"

# Contar frecuencia de cada letra
conteo = Counter(palabra)
print(conteo)  


ordenado = sorted(conteo.items(), key=lambda x: x[1], reverse=True)
print(ordenado) 


ordenado_alfabetico = sorted(conteo.items(), key=lambda x: (-x[1], x[0]))
print(ordenado_alfabetico) 
```

### g. Dada una lista de tuplas, ordenar la primero por el departamento del empleado y luego por la edad.




```python
empleados = [
    ("Maria", "Ventas", 30),
    ("Luis", "Tecnología", 25),
    ("Ana", "Ventas", 25),
    ("Pedro", "Tecnología", 28)
]

# Ordenar primero por departamento, luego por edad (ambos ascendente)
ordenados = sorted(empleados, key=lambda x: (x[1], x[2]))
print(ordenados)
```
