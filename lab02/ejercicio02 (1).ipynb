## Realizar los siguientes ejercicios usando listas de compresión.

### a. Generar una lista con tuplas (n, n^2) para una lista de numeros.

```python
tuplas = [(x,x**2) for x in range(10)]
print(tuplas)
```

### b. Generar una lista con cubos de los números mayores a 10.

```python
cubos_mayores_10 = [x**3 for x in range(50) if x > 10]
print(cubos_mayores_10)
```

### c. Generar una lista con solo los numeros pares.

```python
solo_pares = [x for x in range(20) if x % 2 == 0]
print(solo_pares)
```

### d. Dada la lista numeros, producir: ["valor: 10", "valor: 15", ...]

```python
lista = [1,2,3,4]
lista_nueva = [f"valor:{num}" for num in lista]
print(lista_nueva)
```

### e. Generar una lista con la conversion correspondiente a grados Fahrenheit.

```python
lista = [0,10,20,30,40]
lista_nueva = [f"valor en fahrenheint:{num*1.8 + 32}" for num in lista]
print(lista_nueva)
```

### f. Generar una lista filtrando las palabras de mas de 5 letras.

```python
lista = ["perro","perrote","caballero"]
lista_filtrada = [pal for pal in lista if len(pal) > 5]
print(lista_filtrada)
```

### g. Generar una lista con solo las iniciales de las palabras.

```python
palabras = ["hola", "mundo", "python"]
iniciales = [p[0] for p in palabras]
print(iniciales) 
```

### h. Generar una lista indicando "par" o "impar" dada una lista de numeros.

```python
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
par_impar = ["par" if n % 2 == 0 else "impar" for n in numeros]
print(par_impar)
```

### j. Dada una lista de listas (matriz), aplanala.

```python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
aplanada = [elemento for fila in matriz for elemento in fila]
print(aplanada)  # [1, 2, 3, 4, 5, 6, 7, 8, 9]
```
