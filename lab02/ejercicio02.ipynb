```haskell
--Usa comprensión de listas para

--Obtener los cuadrados de los primeros 10 números.

cuadrado = [ x^2 |x <- [1..10]]
-- cuadrado

--Filtrar los divisibles por 3 entre 1 y 30.
divisibles_3 = [ x |x <- [1..30], mod x 3 == 0]
-- divisibles_3

--Generar una lista de pares (x, y) donde x < y y ambos estén en [1..5]
pares = [(x,y) | x <- [1..5], y <- [1..5], x < y]
--pares

-- Define una función que devuelva los divisores de un número (usando lista por comprensión):

divisores :: Int -> [Int]
divisores n = [ x | x <- [1..n], mod n x == 0 ]
--divisores 15

--Define la función divisoresPropios, crea una función que determine si un número es perfecto (suma de sus divisores propios = número)

divisoresPropios :: Int -> [Int]
divisoresPropios n = [ x | x <- [1..n-1], mod n x == 0 ]


esPerfecto :: Int -> Bool
esPerfecto m = sum (divisoresPropios m) == m
--esPerfecto 6



```
