## Realizar los siguientes ejercicios usando map.

### a. Dada una lista de temperaturas en Celsius, convertirlas a Fahrenheit.

```python
temperaturas = [45,67,34,12,-58]
farenheint = map(lambda t: (t*9/5) + 32, temperaturas)
print(list(farenheint))
```

### b. Dadas dos listas con la misma cantidad de elementos, generar una nueva sumando cada elementos (primer elemento de lista 1 + primer elemento lista 2).

```python
lista1 = [1, 2, 3, 4, 5]
lista2 = [10, 20, 30, 40, 50]

sumas = list(map(lambda x, y: x + y, lista1, lista2))
print(sumas)  
```

### c. Dada una lista de radios de circulos, generar la lista de sus areas.

```python
import math

radios = [1, 2, 3, 4, 5]
areas = list(map(lambda r: math.pi * r**2, radios))
print(areas)
```

### d. Dada un diccionario, generar una lista de tuplas (nombre, valor+10%)

```python
productos = {"pan": 1000, "leche": 2500, "café": 5000}


resultado = list(map(lambda item: (item[0], item[1] * 1.10), productos.items()))
print(resultado)
```

### e. Dada una matriz, multiplicar todos los elementos por 10.

```python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]


matriz_x10 = list(map(lambda fila: list(map(lambda x: x * 10, fila)), matriz))
print(matriz_x10)
```
